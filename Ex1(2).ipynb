```python
import pandas as pd
df=pd.read_csv(r'C:\Users\HI\Downloads\country_wise_latest_modified.csv')
print("Missing values from the list are:\n",df.isnull().sum())
df['Deaths']=df['Deaths'].fillna(df['Deaths'].mean())
df['Recovered']=df['Recovered'].fillna(df['Recovered'].mean())
df['Active']=df['Active'].fillna(df['Active'].mean())
df['New cases']=df['New cases'].fillna(df['New cases'].mean())
df['New deaths']=df['New deaths'].fillna(df['New deaths'].mean())
df['New recovered']=df['New recovered'].fillna(df['New recovered'].mean())
df['Deaths / 100 Cases']=df['Deaths / 100 Cases'].fillna(df['Deaths / 100 Cases'].mean())
df['Recovered / 100 Cases']=df['Recovered / 100 Cases'].fillna(df['Recovered / 100 Cases'].mean())
df['Deaths / 100 Recovered']=df['Deaths / 100 Recovered'].fillna(df['Deaths / 100 Recovered'].mean())
df['1 week change']=df['1 week change'].fillna(df['1 week change'].mean())
df['1 week % increase']=df['1 week % increase'].fillna(df['1 week % increase'].mean())
print("\nMissing values from the updated list are:\n",df.isnull().sum())
```
