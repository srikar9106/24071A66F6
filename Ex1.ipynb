```python
n = int(input("Enter a number: "))
c = 0
for i in range(1, n+1):
  if n%i == 0:
    c = c+1
if c == 2:
  print(n,'is a Prime')
else:
  print(n,'is Non Prime')
```

```python
def fib(n):
  if n==0 or n==1:
    return n
  else:
    return fib(n-1)+fib(n-2)
x=int(input("Enter a number: "))
for i in range(x):
  print(fib(i))

```

```python
def gcd(a,b):
  while b:
    a,b=b,a%b
  return a
def lcm(a,b):
  return int((a*b)/gcd(a,b))
a=int(input('enter 1st number:'))
b=int(input('enter 2nd number:'))
print('GCD:',gcd(a,b))
print('LCM:',lcm(a,b))
```

```python
def add(x,y):
  return x + y

def sub(x,y):
  return x - y

def mult(x,y):
  return x * y

def div(x,y):
  if y==0:
    return "Error in division!"
  else:
    return x / y



def calculator():
  print("Enter the operation:")
  print("1.Add")
  print("2.Subtract")
  print("3.Multiply")
  print("4.Division")
  ops=input()
  a=int(input("Enter 1st number:"))
  b=int(input("Enter 2nd number:"))
  if ops=='1':
    print(f"Result= {add(a,b)}")
  elif ops=='2':
    print(f"Result= {sub(a,b)}")
  elif ops=='3':
    print(f"Result= {mult(a,b)}")
  elif ops=='4':
    print(f"Result= {div(a,b)}")
  else:
    print("Invalid!")

calculator()
```

```python
import random
import string

def generate_password(length=8):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

print(generate_password(5))
```
